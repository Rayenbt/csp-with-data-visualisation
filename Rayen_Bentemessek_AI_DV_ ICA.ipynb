### Scenario 1:

```python
## imported library and initialised the problem
import constraint
problem = constraint.Problem()
```

```python
## added my variables and domains in combinations to avoid redundancy
## example: Ciara Peter Jane is the same as Ciara Jane Peter so that's why I forced these combinations to avoid repeated solutions
## so here my first variable is python, the domain is the possible combinations for python developers (4 persons and 3 roles, whi)
problem.addVariable("Python", [
    ["Ciara", "Peter", "Jane"],  
    ["Ciara", "Peter", "Bruce"],  
    ["Ciara", "Jane", "Bruce"],  
    ["Peter", "Jane", "Bruce"],  
])

problem.addVariable("AI", [
    ["Peter", "Juan"],  
    ["Peter", "Jim"],
    ["Peter", "Maria"], 
    ["Peter", "Anita"],  
    ["Juan", "Jim"],    
    ["Juan", "Anita"],
    ["Juan", "Maria"],
    ["Jim", "Anita"],
    ["Jim", "Maria"],
    ["Anita", "Maria"],
])

problem.addVariable("Web", [
    ["Juan"],  
    ["Mary"],   
    ["Anita"],   
])
problem.addVariable("Database", [
    ["Jim"],  
    ["Jane"],   
])

problem.addVariable("Systems", [
    ["Juan"],  
    ["Jim"],  
    ["Mary"],   
    ["Bruce"],   
])
```

```python
### constraints: 
## -Each employee can only have maximum of two roles 
## -we can only hire a maximum of 4 people
def constraints(Python, AI, Web, Database, Systems):
    teams= [Python, AI, Web, Database, Systems]
    employee_counts = {}
    unique_employees = set()
    ## we iterate through every team, and we update our set of unique employees with the current team (I used a set becaus it doesn't allow duplicates)
    for team in teams:
        unique_employees.update(team)
        unique_employees.discard('Ciara') ## we ignore the existence of ciara
        ## then we iterate through every employee in the team, and we check if he/she is already added, if yes we add 1, if not we add him/her to the dict
        for employee in team:
            if employee in employee_counts:
                employee_counts[employee] += 1
            else:
                employee_counts[employee] = 1

## return True only if all employees counts are less than or equal to 2 AND unique employees are less than or equal to 2
    return max(employee_counts.values()) <= 2 and len(unique_employees) <= 4

problem.addConstraint(constraints,["Python", "AI", "Web", "Database", "Systems"])
```

```python
solutions = problem.getSolutions()
## As there are 2 possible big solutions, and hiring ciara being a soft constraint, I've created 2 lists: 
## One where ciara is included, and the othere where she is not
solutions_with_ciara = []
solutions_without_ciara = []

## I iterated through the found solutions 
for solution in solutions:
    hired_emp = set()
    for team in solution.values():
        for employee in team:
            hired_emp.add(employee)
### If ciara is part of the team, we add the solution to the relevant list, if she's not we add here to the relevant list as well
    if 'Ciara' in hired_emp:
        solutions_with_ciara.append(solution)
    else:
        solutions_without_ciara.append(solution)

## Displayed the total number of unique solutions and then broke it down to the 2 conditions 
##( unique combination of roles not unique set of hired people, further detail of the rationale will be included in the report )
print(f"Total number of unique solutions: {len(solutions)}")
print(f"Total number of solutions where Ciara is part of the team: {len(solutions_with_ciara)}")
print(f"Total number of solutions where Ciara is not part of the team: {len(solutions_without_ciara)}")

print("--------------------------------------------")
print("Solutions where Ciara is part of the team")
print("--------------------------------------------")

teams= ["Python", "AI", "Web", "Database", "Systems"]

## loop will iterate over solutions where ciara is there and make the display readable by detailing each role and the team chosen from the variables
## and domain list
for idx,solution in enumerate(solutions_with_ciara,1):
    print(f"Solution {idx}:")
    for role in teams:
        print(f"{role}: {solution[role]}")
    hired_emp = set()
    for team in solution.values():
        hired_emp.update(team)
        ## we make sure to ignore the existence of Ciara in these solutions
        hired_emp.discard('Ciara')
    print('Hired employees:', hired_emp)
    print("---------")

print("--------------------------------------------")
print("Solutions where Ciara is not part of the team")
print("--------------------------------------------")

for idx,solution in enumerate(solutions_without_ciara,1):
    print(f"Solution {idx}:")
    for role in teams:
        print(f"{role}: {solution[role]}")
    hired_emp = set()
    for team in solution.values():
        hired_emp.update(team)
    print('Hired employees:', hired_emp)
    print("---------")
```

```python
### constraints: very similar to the 1st constraint in scenario 1, the only change is added security to the function's parameters
def constraints(Python, AI, Web, Database, Systems,Security):
    teams= [Python, AI, Web, Database, Systems,Security]
    employee_counts = {}
    unique_employees = set()
    for team in teams:
        unique_employees.update(team)
        for employee in team:
            if employee in employee_counts:
                employee_counts[employee] += 1
            else:
                employee_counts[employee] = 1

    return max(employee_counts.values()) <= 2 and len(unique_employees) <= 5

problem2.addConstraint(constraints,["Python", "AI", "Web", "Database", "Systems","Security"])

```

```python
solutions2 = problem2.getSolutions()
solutions_with_security = []
solutions_without_security = []
## Here I wanted to split the solutions based on our soft constraint: security hired
## same way to what we did with ciara in scenario 1
for solution in solutions2:
    security_hired = solution.get("Security",[])
    if security_hired:
        solutions_with_security.append(solution)
    else:
        solutions_without_security.append(solution)
teams2= ["Python", "AI", "Web", "Database", "Systems","Security"]
print(f"Total number of unique solutions: {len(solutions2)}")
print(f"Total number of solutions where a security employee was hired {len(solutions_with_security)}")
print(f"Total number of solutions where a security employee was not hired: {len(solutions_without_security)}")


print("----------------------------------------------")
print("Solutions where a security employee was hired")
print("----------------------------------------------")
for idx, solution in enumerate(solutions_with_security,1):
    print(f"Solution {idx}:")
    for role in teams2:
        # Each team is printed with the relevant potential employees
        print(f"{role}: {solution[role]}")
    hired_emp = set()
    for team in solution.values():
        hired_emp.update(team)
    print("Hired employees:", hired_emp)
    print("------")

print("---------------------------------------------------")
print("Solutions where a security employee was not hired")
print("---------------------------------------------------")
for idx, solution in enumerate(solutions_without_security,1):
    print(f"Solution {idx}:")
    
    for role in teams2:
        #### Here We check if the security employee slot is empty, we print that it's not assigned 
        if solution[role] == []: 
            print(f"{role}: Not assigned")
        else:
        # Each team is printed with the relevant potential employees
            print(f"{role}: {solution[role]}")
    hired_emp = set()
    for team in solution.values():
        hired_emp.update(team)
    print("Hired employees:", hired_emp)
    print("------")


```

## Solution using DFS Algorithm

```python
### Here I predefined my combinations ( same way in csp )
combinations_scenario_1 = {
    "Python": [
        ["Ciara", "Peter", "Jane"],  
        ["Ciara", "Peter", "Bruce"],  
        ["Ciara", "Jane", "Bruce"],  
        ["Peter", "Jane", "Bruce"],  
    ],
    "AI": [
        ["Peter", "Juan"],  
        ["Peter", "Jim"],
        ["Peter", "Maria"], 
        ["Peter", "Anita"],  
        ["Juan", "Jim"],    
        ["Juan", "Anita"],
        ["Juan", "Maria"],
        ["Jim", "Anita"],
        ["Jim", "Maria"],
        ["Anita", "Maria"],
    ],
    "Web": [
        ["Juan"],  
        ["Mary"],   
        ["Anita"],   
    ],
    "Database": [
        ["Jim"],  
        ["Jane"],   
    ],
    "Systems": [
        ["Juan"],  
        ["Jim"],  
        ["Mary"],   
        ["Bruce"],   
    ]
}

combinations_scenario_2 = {
    "Python": [
        ["Peter", "Jane", "Bruce"],  
    ],
    "AI": [
        ["Peter", "Jim","Anita"], 
        ["Peter", "Jim","Maria"], 
        ["Peter", "Anita","Maria"], 
        ["Jim", "Anita","Maria"],    
    ],
    "Web": [
        ["Anita"],  
        ["Mary"],  
    ],
    "Database": [
        ["Jim"],  
        ["Jane"],   
    ],
    "Systems": [
        ["Bruce"],  
        ["Jim"],    
        ["Mary"],  
    ],
    "Security": [
        ["Mary"],
        ["Maria"],
        []
    ]
}
```

```python
## I created two function that we will need later in order to get the number of employees
## The approach so far is very similar to what I used in CSP
## scenario 1: if ciara is hired , it becomes 5 = Ciara + 4, if she's not we only need 4
def scenario_1(employee_counts):
    unique_employees = set(employee_counts.keys())
    ### This will check if each employee is assigned a maximum of two roles
    for employee, count in employee_counts.items():
        if count > 2 :
            return False
    ## then we will exclude ciara from the unique employees, that has to be 4 or less
    if 'Ciara' in unique_employees:
        unique_employees.remove('Ciara')
    return len(unique_employees) <=4


## scenario 2: using the additional restrictions, we will require one security employee, and ciara and juan are partners
## so the number of unique employees need to be less than or 5
def scenario_2(employee_counts):
    unique_employees = set(employee_counts.keys())
     ### This will check if each employee is assigned a maximum of two roles
    for employee, count in employee_counts.items():
        if count > 2 :
            return False

    return len(unique_employees) <= 5
```

```python
## This function will perform the dfs algorithm in order to find the valid solutions , using the functions in the previous cell
## the function has 2 parameters:
## combination_scenario which is the dictionary of possible combinations by roles, and the scenario treated because we have 2
def dfs(combination_scenario, scenario):
    ## we start by a list of roles from combination_scenario
    roles = list(combination_scenario.keys())
    solutions = []
    ## initial state would start with no roles assigned yet
    initial_state = (0,{},{})
    ## we start the stack first with the initial state
    stack = [initial_state]

    while stack:
        ## the index is currently 0, current_solution is empty, and employee_counts is null, 
        ## but we cannot put it as 0 because it's a dictionary showing each employee (key) with his value (number of roles occupied)
        index, current_solution, employee_counts = stack.pop()

        ## this checks if index is equal to the number of roles, tests the solutions with the other functions, if all is good, we got a new solution 
        if index == len(roles):
            if scenario == 1 and scenario_1(employee_counts):
                solutions.append(dict(current_solution))
            elif scenario == 2 and scenario_2(employee_counts):
                solutions.append(dict(current_solution))
            continue
        
        role = roles[index]
        ## now we loop through all the combinations for the range of our roles (based on the index)
        for combination in combination_scenario[role]:
            ## we dont want our real employee_counts to disappear so we create a copy to do the job
            copy_employee_counts = dict(employee_counts)

            ## now we loop through every employee in the role combination, example: Jane in Python
            for c in combination:
                ## we update the number of times jane was hired because she can have more than one role, so if it's already there, we get
                ## its value "c" and we add 1, if not we make it 1
                copy_employee_counts[c] = copy_employee_counts.get(c,0)+1
                
            ## once we passed the conditions, solution should be fine ( we still need to double check it once the index meets the number of roles condition)
            ## we assign the current solution to a new variable that will be added to the next state
            new_solution = dict(current_solution)
            ## we add the current combination to the solution
            new_solution[role] = combination
            ## we move to the next state, once index reaches the len(roles), we execute the first if condition in the while loop, if it passes, we got a full solution
            next_state = (index + 1, new_solution,copy_employee_counts)
            stack.append(next_state)
       
    return solutions

```

```python
## we create list of roles names from the combination dictionary keys (i.e Python, AI etc..)
teams1= list(combinations_scenario_1.keys())
teams2= list(combinations_scenario_2.keys())


## I've used the same type and logic of output as in the csp
print("=== Scenario 1 Solutions ===")
solutions1 = dfs(combinations_scenario_1, scenario=1)
print("Number of solutions for Scenario 1:", len(solutions1))
for idx, solution in enumerate(solutions1, 1):
    print(f"Solution {idx}:")
    for role in teams1:
        print(f"{role}: {solution[role]}")
    hired_emp = set()
    for team in solution.values():
        hired_emp.update(team)
        hired_emp.discard('Ciara')
    print('Hired employees:', hired_emp)
    print("---------")

print("\n=== Scenario 2 Solutions ===")
solutions2 = dfs(combinations_scenario_2, scenario=2)
print("Number of solutions for Scenario 2:", len(solutions2))
for idx, solution in enumerate(solutions2,1):
    print(f"Solution {idx}:")
    for role in teams2:
        if solution[role] == []: 
            print(f"{role}: Not assigned")
        else:
            print(f"{role}: {solution[role]}")
    hired_emp = set()
    for team in solution.values():
        hired_emp.update(team)
    print("Hired employees:", hired_emp)
    print("------")
```

```python
## In order to do the interactive visualisation, I chose to work with plotly library, more specifically dash framework
## further details and explanations of choices will be in the report
import dash
from dash import dcc, html, Dash
from dash.dependencies import Input, Output
import plotly.graph_objects as go


## this function will introduce the layout for the dropdown options, we will need it later on and for the GUI as well
def get_layout():
    return html.Div([
        ## There will be a scenario selector (dropdown) where the user chooses which scenario they want to explore
        dcc.Dropdown(
            id='scenario-selector',
            options=[
                {'label': 'Scenario 1', 'value': 'Scenario 1'},
                {'label': 'Scenario 2', 'value': 'Scenario 2'}
            ],
            ## default value will be scenario 1, to avoid blank charts when there is no choice made
            value='Scenario 1',
            clearable=False
        ),
        ## same logic for the group and solution selectors, but I have got more options there with different values so I have created functions for that
        
        dcc.Dropdown(
            id='group-selector',
            options=[],
            value=None,
            clearable=False
        ),
        dcc.Dropdown(
            id='solution-selector',
            options=[],
            value=None,
            clearable=False
        ),
        dcc.Graph(id='sankey-diagram')
    ])

## this function is responsible for registering the callbacks and creating the sankey diagram
def sankey_diagram(app, solutions):
    ## this callback will update group-selector options and values with the corresponding values after running
    @app.callback(
        Output('group-selector', 'options'),
        Output('group-selector', 'value'),
        Input('scenario-selector', 'value')
    )
    
    ## the options and values will be taken from this function that splits each scenario into 2 groups
    ## scenario 1, group 1 where ciara is hired, group 2 where solutions where ciara is not hired
    ## scenario 2 will be split based on the presence/absence of security employee
    def update_group_dropdown(scenario):
        if scenario == 'Scenario 1':
            options = [
                {'label': 'Solutions with Ciara', 'value': 'Solutions with Ciara'},
                {'label': 'Solutions without Ciara', 'value': 'Solutions without Ciara'},
            ]
            default_value = 'Solutions with Ciara'
        else:  ## we only have 2 scenarios
            options = [
                {'label': 'Solutions with Security', 'value': 'Solutions with Security'},
                {'label': 'Solutions without Security', 'value': 'Solutions without Security'},
            ]
            default_value = 'Solutions with Security'
            ## then we will return the options and the default_values 
        return options, default_value
    
    
    ## same logic but this callback is based on the solutions now
    ## it will take scenario and group and updates the solutions with their values
    @app.callback(
        [Output('solution-selector', 'options'),
         Output('solution-selector', 'value')],
        [Input('scenario-selector', 'value'),
         Input('group-selector', 'value')]
    )
    
    
    def update_solution_dropdown(scenario, group):
        ## here we take take the relevant solutions based on the scenario and group
        solutions_map = solutions.get(scenario,{})
        solutions_group = solutions_map.get(group,[])
        ## then we create the options for the dropdown menu
        options = [{'label': f'Solution {i+1}', 'value': i} for i in range(len(solutions_group))]
        ## default value is always the first
        default_value = options[0]['value']
        return options, default_value
    
    ## now this callback will update the sankey diagram from the scenario group and solution
    @app.callback(
        Output('sankey-diagram', 'figure'),
        [Input('scenario-selector', 'value'),
         Input('group-selector', 'value'),
         Input('solution-selector', 'value')]
    )
    def update_sankey(scenario, group, solution_index):
        solutions_map = solutions.get(scenario, {})
        solution = solutions_map.get(group, [])[solution_index]
        
        ## roles will take the keys from the solution
        roles = list(solution.keys())
        ## and then we get the unique employees
        unique_employees = set()
        for employees in solution.values():
            for employee in employees:
                unique_employees.add(employee)
        employees = list(unique_employees)
        ## then we add roles and employees to a list of labels which will be the nodes of the Sankey diagram
        labels = employees + roles
    
        ## then we create the sankey diagram links, we will need a source, target and value
        source, target, value = [], [], []
        for role, employees in solution.items():
            ## we need to keep the distribution of employees as it is
            for employee in employees:
                source.append(labels.index(employee)) ## source will be employee
                target.append(labels.index(role)) ## target is the role
                value.append(1) ## value is 1 for each link, example: jane to database the value is 1, jane to python is 1
        ## Now we create the sankey diagram
        fig = go.Figure(data=[go.Sankey(
            node=dict(label=labels, pad=30, thickness=20), ## our nodes will be the labels ( employees + roles )
            link=dict(source=source, target=target, value=value) ## links will start from source 
        )])
        fig.update_layout(title_text="Sankey Diagram", font_size=15)
        return fig

solutions = {
    'Scenario 1': {
        'Solutions with Ciara': solutions_with_ciara,
        'Solutions without Ciara': solutions_without_ciara
    },
    'Scenario 2': {
        'Solutions with Security': solutions_with_security,
        'Solutions without Security': solutions_without_security
    }
}
app = Dash(__name__)
app.layout = get_layout()
sankey_diagram(app, solutions)


if __name__ == "__main__":
    app.run_server(debug=False, port=8050)
```
